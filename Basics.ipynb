# Read the data 

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
print(star_wars.columns)
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

# converting each column to a Boolean values

```python
yes_no = {"Yes": True, "No": False}
star_war =   star_wars.copy()
cols = ['Have you seen any of the 6 films in the Star Wars franchise?'
, 'Do you consider yourself to be a fan of the Star Wars film franchise?']
for c in cols :
        star_war[c] = star_war[c].map(yes_no).copy()
star_war.head()        
```

# convert the columns to a Boolean & rename 

```python
bool_movies = {'Star Wars: Episode I  The Phantom Menace': True,
               "Star Wars: Episode II  Attack of the Clones": True,
               "Star Wars: Episode III  Revenge of the Sith": True,
               "Star Wars: Episode IV  A New Hope": True,
               "Star Wars: Episode V The Empire Strikes Back": True,
               "Star Wars: Episode VI Return of the Jedi": True,
               np.nan :False
}
for col in star_war.columns[3:9]:
    star_war[col] = star_war[col].map(bool_movies)
```

```python
star_war = star_war.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1',
                                    'Unnamed: 4':'seen_2',
                                    'Unnamed: 5':'seen_3',
                                    'Unnamed: 6':'seen_4',
                                    'Unnamed: 7':'seen_5',
                                    'Unnamed: 8':'seen_6' 
})
star_war.head()
```

# convert columns to a numeric 

```python
star_war[star_war.columns[9:15]] = star_war[star_war.columns[9:15]].astype(float)
```

```python
star_war = star_war.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
                         'Unnamed: 10': 'ranking_2',
                         'Unnamed: 11': 'ranking_3',
                         'Unnamed: 12': 'ranking_4',
                         'Unnamed: 13': 'ranking_5',
                         'Unnamed: 14': 'ranking_6'
     
})
```
